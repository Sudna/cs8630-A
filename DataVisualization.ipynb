```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt

def read_txt_files(folder_path):
  txt_files = glob.glob(os.path.join(folder_path, '*.txt'))
  for file_path in txt_files:
    with open(file_path, 'r') as file:
      print(f"Reading file: {file_path}")
      lines = file.readlines()

      # Find the line containing "Wavelength nm."
      start_index = None
      for i, line in enumerate(lines):
        #print(f"Line {i}: {line}")
        if "Wavelength nm." in line:
          start_index = i + 1
          break

      if start_index is not None:
        wavelength = []
        intensity = []
        for line in lines[start_index:]:
          values = line.strip().split(',')
          #print(f"Values: {values}")
          if len(values) >= 2:
            wavelength.append(float(values[0]))
            intensity.append(float(values[1]))
        print(f"Wavelength: {wavelength}")
        print(f"Intensity: {intensity}")
        wave = np.array(wavelength)
        intens = np.array(intensity)
        plt.plot(wave, intens)



data_folder = '/content/drive/MyDrive/cs8630-A/Homework/data'
read_txt_files(data_folder)
```
